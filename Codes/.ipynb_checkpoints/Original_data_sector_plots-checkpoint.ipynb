```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
sectors = {
    # Healthcare
    'JNJ': 'Healthcare', # Johnson & Johnson
    'MRK': 'Healthcare', # Merck & Co.
    'PFE': 'Healthcare', # Pfizer
    'UNH': 'Healthcare', # UnitedHealth
    'AMGN': 'Healthcare', #Amgen

    # Finance
    'GS': 'Finance', # Goldman Sach
    'AXP': 'Finance', # American Express
    'JPM': 'Finance', # JPMorgan 
    'TRV': 'Finance', # Travelers Companies

    # Telecom
    'VZ': 'Telecom', # Verizon

    # Consumer Retail
    'KO': 'Consumer Retail', # Coca-cola
    'MCD': 'Consumer Retail', # McDonalds
    'WMT': 'Consumer Retail', # Walmart
    'HD': 'Consumer Retail', # Home Depot
    'NKE': 'Consumer Retail', # Nike
    'DIS': 'Consumer Retail', # Walt Disney
    'PG': 'Consumer Retail', # Proctor & Gamble
    'V': 'Consumer Retail', # Visa

    # Materials
    'MMM': 'Materials', # 3M Co.
    'DOW': 'Materials', # Chemicals
    'CAT': 'Materials', # Caterpillar
    'CVX': 'Materials', # Chevron
    'XOM': 'Materials', # Exxon
    'BA': 'Materials', # Boeing
    'INTC': 'Materials',# Intel
    'HON': 'Materials', # HoneyWell (Aerospace)

    # Technology
    'AAPL': 'Technology', # Apple
    'MSFT': 'Technology', # Microsoft
    'CSCO': 'Technology', # Cisco
    'CRM': 'Technology', # Salesforce
}

sector_colors = {
    'Healthcare': 'green',
    'Finance': 'blue',
    'Telecom': 'red',
    'Consumer Retail': 'purple',
    'Materials': 'orange',
    'Technology': 'yellow',
}
```

```python
df = pd.read_csv('../Datasets/dow.csv', index_col=0)
```

```python
# Make sure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# Map each row to a sector based on the 'tic' column
df['sector'] = df['tic'].map(sectors)
```

```python
# Calculate the average closing price for each sector on each day
sector_daily_close = df.groupby(['date', 'sector'])['close'].mean().unstack()
```

```python
# Plot each sector's average closing price over time
for sector in sector_daily_close.columns:
    plt.plot(sector_daily_close.index, sector_daily_close[sector], label=sector, color=sector_colors.get(sector, 'black'))

# Improve date formatting and visibility
plt.gca().xaxis.set_major_locator(mdates.YearLocator())   # Set major ticks to yearly intervals
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())  # Set minor ticks to monthly intervals
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Show only the year

plt.gcf().autofmt_xdate()  # Automatically format x-axis labels to fit into the plot
plt.title('Average Closing Price by Sector')
plt.xlabel('Date')
plt.ylabel('Average Closing Price')
plt.legend(title='Sector')
plt.tight_layout()  # Adjust layout to fit everything neatly
plt.show()
```

```python
# Filter the DataFrame for the date range 2019 to 2023
start_date = "2018-01-01"
end_date = "2023-02-28"
mask = (df['date'] >= start_date) & (df['date'] <= end_date)
filtered_df = df.loc[mask]

# Proceed with grouping by date and sector if you haven't already
sector_daily_close = filtered_df.groupby(['date', 'sector'])['close'].mean().unstack()
```

```python
import os

# Define the directory where you want to save the plot
plot_directory = '../Plots'
# Ensure the directory exists
os.makedirs(plot_directory, exist_ok=True)

# Plot each sector's average closing price over time
for sector in sector_daily_close.columns:
    plt.plot(sector_daily_close.index, sector_daily_close[sector], label=sector, color=sector_colors.get(sector, 'black'))

# Improve date formatting and visibility
plt.gca().xaxis.set_major_locator(mdates.YearLocator())   # Set major ticks to yearly intervals
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())  # Set minor ticks to monthly intervals
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Show only the year

plt.gcf().autofmt_xdate()  # Automatically format x-axis labels to fit into the plot
plt.title('Average Closing Price by Sector')
plt.xlabel('Date')
plt.ylabel('Average Closing Price')
plt.legend(title='Sector')
plt.tight_layout()  # Adjust layout to fit everything neatly

# Specify the full path for the file where you want to save the plot
plot_path = os.path.join(plot_directory, 'Original_Average_Closing_Price_by_Sector_Trade.png')

# Save the plot to the file
plt.savefig(plot_path)

plt.show()
```
